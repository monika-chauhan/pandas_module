```python
# Load the CSV file 
import pandas as pd 
df = pd.read_csv('Pokemon.csv')
df
```

```python
# Find the columns of Dataframe
df.columns
```

```python
# Filter the Data based on given Data 
df.loc[df['Type 1'] == 'Grass']
```

```python
# Describe the DataFrame Columns 
df.describe()
```

```python
# Sorting the Value based on the passing columns and order 
df.sort_values(['Name','HP'],ascending=[1,0])
```

```python
# Filter Data 
#filter_data = df.loc[df['Type 1'] == ('Grass')]

# Filter with and condition 
#filter_data = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')] 

# Filter with OR condition
filter_data = df.loc[(df['Type 1'] == 'Grass') | (df['Type 2'] == 'Poison')]

filter_data
```

```python
# Reset Index 
filter_data_index = filter_data.reset_index(drop=True)
filter_data_index
```

```python
# Filter Data with Name contains 
#filter_name = df.loc[df['Name'].str.contains('Mega')]

# Passing multiple name 
import re
#filter_name = df.loc[df['Name'].str.contains('Mega | Char', regex = True )]

# Passing Regukar Expression 
filter_name = df.loc[df['Name'].str.contains('^Pi[a-z]*',regex=True)] 

filter_name
```

```python
# conditon Changes 
#df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flamer'
df.loc[df['Type 1'] == 'Flamer', 'Type 1'] = 'Fire'
df
```

```python
# conditon base updation of value 
#df.loc[df['Total'] > 500, ['Generation','Legendary']] = 'Test Value'
df.loc[df['Total'] > 500, ['Generation','Legendary']] = [1,False]
df
```

```python
# Aggregate Statistics (Group by) 
df.sort_values('HP').groupby(['Type 1','Type 2'])
df
```

```python
# calculating the sum based on Type 1 using group by clause
s=df.groupby('Type 1').sum()
s
```

```python
# add the count column in Data Frame and update the count value based on the Count function
df['count'] = 1
count = df.groupby('Type 1').count()['count']
count
```
